```python
from docx import Document
import pandas as pd
import os
```

```python
#Set directory where document ouputs will be printed
working_dir ='G:\Student_Survey_Analysis\AnalysisResults\\'
os.chdir(working_dir)
```

```python
data_path ="UZIMA Student College Survey.csv" 
# Load the data
student_survey_data = pd.read_csv(data_path)

# Drop the first two rows
student_survey_data = student_survey_data.drop([0, 1])
student_survey_data = student_survey_data.reset_index(drop=True)
```

```python
#Count number of particioants who have consented using Consent3
# Ensure startdate is in datetime format
student_survey_data['StartDate'] = pd.to_datetime(student_survey_data['StartDate'])
student_survey_data['EndDate'] = pd.to_datetime(student_survey_data['EndDate'])

# Extract year and month
student_survey_data['Year'] = student_survey_data['StartDate'].dt.year
student_survey_data['Month'] = student_survey_data['StartDate'].dt.month

# convertConsent3 to numeric
student_survey_data['Consent3']= pd.to_numeric(student_survey_data['Consent3'], errors='coerce')

# count_check_df = student_survey_data[['ResponseID','StartDate','Consent3','Year','Month']]
consented_participants = student_survey_data[student_survey_data['Consent3'] ==1]
# consented_participants.to_csv("consented.csv")
# count_check_df.to_csv("count.csv")
# # Group by year and month, and count records
consented_participants_per_period = consented_participants.groupby(['Year', 'Month']).size().reset_index(name='Count')
consented_participants_per_period
total_count = consented_participants_per_period['Count'].sum()

# # Create a Word document
doc = Document()

# Add a title
doc.add_heading('Count of Records For Participants who Consented Grouped by Year and Month', level=1)

# Add the table
table = doc.add_table(rows=1, cols=3)
table.style = 'Table Grid'

# Define the table header
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Year'
hdr_cells[1].text = 'Month'
hdr_cells[2].text = 'Count'

# Add rows to the table
for index, row in consented_participants_per_period.iterrows():
    row_cells = table.add_row().cells
    row_cells[0].text = str(row['Year'])
    row_cells[1].text = str(row['Month'])
    row_cells[2].text = str(row['Count'])
# Add a row for the total count
total_row_cells = table.add_row().cells
total_row_cells[0].text = ''
total_row_cells[1].text = 'Total'
total_row_cells[2].text = str(total_count)

doc.save('Consent_Counts_Grouped_by_Year_and_Month.docx')
```

```python
#Function To calculate NAN for variables with no branching
def create_nan_summaries(dataframe):
    #get the name of the dataframe
    #Generate NAN count:
    doc = Document()
    # Add a heading to the document
    doc.add_heading('DataFrame Missing Values Report', level=1)

    #Add table with column names and missing counts
    table = doc.add_table(rows=1, cols=2)
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Column Name'
    hdr_cells[1].text = 'Missing Count'

    for cols in dataframe:
        na_count = dataframe[cols].isna().sum()
        row_cells = table.add_row().cells
        row_cells[0].text = cols
        row_cells[1].text = str(na_count)   
 
    dfName=dataframe.name
    folder_path="QualityReports//"
    doc.save(os.path.join(folder_path, f'{dfName}_missing_data.docx'))
```

```python
#Call Function to create the NAN summaries for each section with no branching
create_nan_summaries(background_df)
create_nan_summaries(your_health_df)
create_nan_summaries(sexuality_df)
```
